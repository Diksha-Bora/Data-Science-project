MULTI-LINEAR REGRESSION

```python
import numpy as np 
import pandas as pd 
# y = m1x1 + mx2 + ....+c
x1 = np.random.rand(200)
x2 = 2*np.random.rand(200) +4
y = 2*x1+3*x2+9+np.random.rand(200)
print(y)
```

```python
import matplotlib.pyplot as plt
plt.scatter(y,x1,label="X1")
plt.scatter(y,x2,label="X2")
plt.legend()
plt.show()
```

```python
data = pd.DataFrame({'x1':x1,'x2':x2,'y_actual':y})
data.head()
```

```python
input = data.drop('y_actual',axis=1)
output = data['y_actual']
```

```python
# MULTI-LINEAR REGRESSION MODEL
from sklearn.linear_model import LinearRegression
mul_reg = LinearRegression()
mul_reg.fit(input,output)
```

```python
print("The coefficeient are: ",mul_reg.coef_)
print("The intercept is ",mul_reg.intercept_)
```

```python
# PREDICTION USING MULTI-LINEAR REGRESSION MODEL
y_pred = mul_reg.predict(input)
y_pred
```

```python
# CALCULATION OF METRICS - how the model performed
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,r2_score
print("The mean_absolute_error is : ",mean_absolute_error(output,y_pred))
print("The mean_absolute_percentage_error is : ",mean_absolute_percentage_error(output,y_pred))
print("The mean_squared_error is : ",mean_squared_error(output,y_pred))
print("The r2_score is : ",r2_score(output,y_pred))
```

```python
import seaborn as sns
print(sns.scatterplot(data = data,x = output,y = y_pred,hue="x1"))
```

```python
print(sns.scatterplot(data = data,x = output,y = y_pred,hue="x2"))
```

MULTI-LINEAR REGRESSION USING DEGREE = 3


```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

# DATA GENERATION

x1 = 2*np.random.rand(200)+np.random.rand(200)+2
x2 = 4*np.random.rand(200)+np.random.rand(200)+3
y = x1**3 + 4*x2**2 + 12
print(y)
```

```python
# REPRESENTING RELATION BETWEEN X1,X2,Y
plt.scatter(y, x1, color='b', label='X1')
plt.scatter(y, x2, color='g', label='X2')
plt.legend()
plt.show()
```

```python
data = pd.DataFrame({'x1':x1,'x2':x2,'y_actual':y})
data.head()
```

```python
input = data.drop('y_actual', axis=1)
output = data['y_actual']
```

```python
# USING LIBRARY FOR CREATING POLYNOMIAL INPUT FOR DEGREE=3
from sklearn.preprocessing import PolynomialFeatures
poly_feat = PolynomialFeatures(degree=3)
poly_input = poly_feat.fit_transform(input)
poly_input
```

```python
# TRAINING MODEL
from sklearn.linear_model import LinearRegression
mul_lin_reg = LinearRegression()
mul_lin_reg.fit(poly_input,output)
```

```python
print("The coefficient are :",mul_lin_reg.coef_)
print("The intercept is :",mul_lin_reg.intercept_)
```

```python
y_pred = mul_lin_reg.predict(poly_input)
y_pred[0]
```

```python
# CALCULATION OF METRICS - how the model performed
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,r2_score
print("The mean_absolute_error is : ",mean_absolute_error(output,y_pred))
print("The mean_absolute_percentage_error is : ",mean_absolute_percentage_error(output,y_pred))
print("The mean_squared_error is : ",mean_squared_error(output,y_pred))
print("The r2_score is : ",r2_score(output,y_pred))
```

```python
# REPRESENTATION OF RELATION BETWEEN ACTUAL AND PREDICTED OUTPUT
plt.scatter(y_pred,output)
```
