MULTI-LINEAR REGRESSION

```python
import numpy as np 
import pandas as pd 
# y = m1x1 + mx2 + ....+c
x1 = np.random.rand(200)
x2 = 2*np.random.rand(200) +4
y = 2*x1+3*x2+9+np.random.rand(200)
print(y)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x1,x2,y)
```

```python
data = pd.DataFrame({'x1':x1,'x2':x2,'y_actual':y})
data.head()
```

```python
input = data.drop('y_actual',axis=1)
output = data['y_actual']
```

```python
# MULTI-LINEAR REGRESSION MODEL
from sklearn.linear_model import LinearRegression
mul_reg = LinearRegression()
mul_reg.fit(input,output)
```

```python
print("The coefficeient are: ",mul_reg.coef_)
print("The intercept is ",mul_reg.intercept_)
```

```python
# PREDICTION USING MULTI-LINEAR REGRESSION MODEL
y_pred = mul_reg.predict(input)
y_pred
```

```python
# CALCULATION OF METRICS - how the model performed
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,r2_score
print("The mean_absolute_error is : ",mean_absolute_error(output,y_pred))
print("The mean_absolute_percentage_error is : ",mean_absolute_percentage_error(output,y_pred))
print("The mean_squared_error is : ",mean_squared_error(output,y_pred))
print("The r2_score is : ",r2_score(output,y_pred))
```

```python
import seaborn as sns
print(sns.scatterplot(data = data,x = output,y = y_pred,hue="x1"))
```
