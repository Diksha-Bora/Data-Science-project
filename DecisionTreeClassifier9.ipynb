```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x1 = np.random.randint(100,size=(20))
x2 = np.random.randint(100,size=(20))
y = np.random.randint(3,size=(20))
```

```python
data = pd.DataFrame({"x1":x1,"x2":x2,"y":y})
data.head()
```

```python
input = data.drop("y",axis = 1)
output =data["y"]
```

```python
from sklearn.model_selection import train_test_split
input_train,input_test,output_train,output_test=train_test_split(input,output,test_size=0.2,random_state=45,stratify=data["y"])
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(input_train,output_train)
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=[25,12])
plot_tree(dtc)
plt.show()
```

```python
y_pred_train = dtc.predict(input_train)
y_pred_test = dtc.predict(input_test)
```

```python
from sklearn.metrics import classification_report
print("TRAIN\n")
print(classification_report(output_train,y_pred_train))
print("TEST\n")
print(classification_report(output_test,y_pred_test))
```
