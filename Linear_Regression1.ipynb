```python
# Importing required libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
# creating data
x = np.random.rand(10)
print(x)
print(x.shape)
y = x = np.random.rand(10)
# bar graph
plt.bar(x,y,color=['yellow','pink'])
plt.xlabel("how funny you can be")
plt.ylabel("how funny I'm",color='brown')
plt.title="hehehe"
plt.show()
```

```python
# Creating dataset to perfoem Linear Regression
x = np.random.rand(500,)+2*np.random.rand(500,)
y = 2*x + 5+np.random.rand(500,)
plt.scatter(x,y)
```

```python
data= pd.DataFrame({'x':x,'y_actual':y})
data.head()
```

```python
import seaborn as sns
sns.boxplot(data = data)
```

```python
data.info()
```

```python
# Linear Regression Model
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
input = data[['x']]
output= data['y_actual']
lin_reg.fit(input, output)
```

```python
# Predicting data using Linear Regression
y_pred = lin_reg.predict(input)
y_pred
```

```python
plt.scatter(y_pred,output)
```

```python
print('weights',lin_reg.coef_)
print('intercept',lin_reg.intercept_)
```

```python
# Calculation of how the model performed
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error,r2_score
print("Calculating the matrics")
print("MEAN-ABSOLUTE-ERROR is ",mean_absolute_error(y_pred,output))
print("MEAN-ABSOLUTE-PERCENTAGE-ERROR is ",mean_absolute_percentage_error(output,y_pred))
print("MEAN-SQUARED-ERROR is ",mean_squared_error(output,y_pred))
print("R2-SQAURE is ",r2_score(output,y_pred))
```
