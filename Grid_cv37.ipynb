GRID CV FOR RANDOM FOREST CLASSIFIER

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
b_cancer = load_breast_cancer()
data = pd.DataFrame(b_cancer.data , columns=b_cancer.feature_names)
data["target"] = b_cancer.target
data.head()
```

```python
x = data.drop("target",axis =1 )
y = data["target"]
```

```python
parameters = {'criterion':("gini", "entropy", "log_loss"),
              'max_depth':(5,7,8,10),
              'bootstrap':(True, False),
              'random_state':(45,12,85,35),
              'warm_start':(True, False)}
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
gs = GridSearchCV(estimator= rf , param_grid = parameters)
```

```python
gs.fit(x,y)
```

```python
gs.cv_results_
```

```python
gs.best_estimator_
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=45, stratify =y)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf =RandomForestClassifier(bootstrap=False, criterion='entropy', max_depth=7,
                       random_state=85, warm_start=True)
rf.fit(x_train,y_train)
```

```python
y_pred_train = rf.predict(x_train)
y_pred_test = rf.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print("TRAIN\n")
print("mean_absolute_error is :\n",classification_report(y_train, y_pred_train))
print("mean_absolute_percentage_error is :\n",confusion_matrix(y_train, y_pred_train))
print("TEST\n")
print("mean_absolute_error is :\n",classification_report(y_test, y_pred_test))
print("mean_absolute_percentage_error is :\n",confusion_matrix(y_test, y_pred_test))
```
