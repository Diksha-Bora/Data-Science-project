```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
data = pd.DataFrame(diabetes.data , columns = diabetes.feature_names)
data["target"] = diabetes.target
data.head()
```

```python
x = data.drop("target", axis =1)
y = data["target"]
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
```

```python
parameters = {'criterion':("squared_error", "absolute_error", "friedman_mse", "poisson"),
             'max_depth':(5,8,7,3),
             'bootstrap':(True,False),
             'random_state':(52,5,85,45),
             'warm_start':(True,False)}
```

```python
from sklearn.model_selection import GridSearchCV
model = GridSearchCV(estimator= rf , param_grid= parameters)
```

```python
model.fit(x,y)
```

```python
model.cv_results_
```

```python
model.best_estimator_
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=45)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf =RandomForestRegressor(max_depth=5, random_state=85, warm_start=True)
rf.fit(x_train,y_train)
```

```python
y_pred_train = rf.predict(x_train)
y_pred_test = rf.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, r2_score
print("TRAIN\n")
print("mean_absolute_error is :",mean_absolute_error(y_train, y_pred_train))
print("mean_absolute_percentage_error is :",mean_absolute_percentage_error(y_train, y_pred_train))
print("r2_score is :",r2_score(y_train, y_pred_train))
print("TEST\n")
print("mean_absolute_error is :",mean_absolute_error(y_test, y_pred_test))
print("mean_absolute_percentage_error is :",mean_absolute_percentage_error(y_test, y_pred_test))
print("r2_score is :",r2_score(y_test, y_pred_test))
```
