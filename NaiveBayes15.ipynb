# **Other oversampling techniques and their f1_score**

---



```python
# IMPORTING REQUIRED LIBRARIES
import numpy as p 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
# IMPORTING DATASET
data = pd.read_csv("/content/adult.csv",header=None)
data.head()
```

```python
# RENAMING COLUMN NAMES
new_col = []
for i in range(len(data.columns)):
  new_col.append("col_"+str(i))
data.columns = new_col  
data.head()
```

```python
data.info()
```

```python
# ANALYZING DATA THAT HAS NO OR LESS INFERENCE TO TARGET_COLUMN
print("col_10")
print(data["col_10"].value_counts())
print("\ncol_11")
print(data["col_11"].value_counts())
```

```python
# REMOVING DATA THAT HAS NO OR LESS INFERENCE TO TARGET_COLUMN
data = data.drop(["col_10","col_11"],axis=1)
```

```python
# CATEGORIZING DATA BASED ON ITS DATATYPE
cat_col = []
num_col = []
for i in data.columns:
  if(data[i].dtype=="object"):
    cat_col.append(i)
  else:
    num_col.append(i)
print("cat_col:",cat_col) 
print("num_col:",num_col)     
```

```python
# LABEL ENCODING CATEGORICAL COLUMN
from sklearn.preprocessing import LabelEncoder
for i in data.columns:
 le = LabelEncoder()
 encoded_col = le.fit_transform(data[i]) 
 data[i+"_encoded"] = encoded_col
data.head() 
```

```python
data = data.drop(cat_col,axis=1)
data.head()
```

```python
# CREATING DATA INTO X & Y FOR MODEL TRAINING 
x = data.drop("col_14_encoded",axis=1)
y = data["col_14_encoded"]
```

```python
# TRAIN-TEST-SPLIT
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=52)
```

RANDOM OVER SAMPLER

```python
# OVERSAMPLING DATA FOR BETTER PREDICTION OF TARGET-COLUMN 
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(sampling_strategy=0.95)
x_train_overs1,y_train_overs1 = ros.fit_resample(x_train,y_train)
```

```python
# MODEL SELECTION AND TRAINING 
from sklearn.naive_bayes import GaussianNB 
gsb = GaussianNB()
gsb.fit(x_train_overs1,y_train_overs1)
```

```python
# PREDICTION USING MODEL 
y_pred_train = gsb.predict(x_train_overs1)
y_pred_test = gsb.predict(x_test)
```

```python
# MODEL EVALUATION
from sklearn.metrics import f1_score 
print("RANDOM OVERSAMPLER")
print("TRAIN :",f1_score(y_train_overs1,y_pred_train))
print("TEST :",f1_score(y_test,y_pred_test))
```

ADASYN OVER-SAMPLING TECHNIQUE

```python
# OVERSAMPLING DATA FOR BETTER PREDICTION OF TARGET-COLUMN 
from imblearn.over_sampling import ADASYN
ad = ADASYN(sampling_strategy=0.95)
x_train_overs2,y_train_overs2 = ad.fit_resample(x_train,y_train)
```

```python
# MODEL SELECTION AND TRAINING AND PREDICTION USING MODEL
from sklearn.naive_bayes import GaussianNB 
gsb = GaussianNB()
gsb.fit(x_train_overs2,y_train_overs2)
y_pred_train = gsb.predict(x_train_overs2)
y_pred_test = gsb.predict(x_test)
```

```python
# MODEL EVALUATION
from sklearn.metrics import f1_score 
print("ADASYN OVERSAMPLER")
print("TRAIN :",f1_score(y_train_overs2,y_pred_train))
print("TEST :",f1_score(y_test,y_pred_test))
```

SMOTE OVER-SAMPLING 

```python
# OVERSAMPLING DATA FOR BETTER PREDICTION OF TARGET-COLUMN 
from imblearn.over_sampling import SMOTE
sm = SMOTE(sampling_strategy=0.95)
x_train_overs3,y_train_overs3 = sm.fit_resample(x_train,y_train)
```

```python
# MODEL SELECTION AND TRAINING AND PREDICTION USING MODEL
from sklearn.naive_bayes import GaussianNB 
gsb = GaussianNB()
gsb.fit(x_train_overs3,y_train_overs3)
y_pred_train = gsb.predict(x_train_overs3)
y_pred_test = gsb.predict(x_test)
```

```python
# MODEL EVALUATION
from sklearn.metrics import f1_score 
print("SMOTE OVERSAMPLER")
print("TRAIN :",f1_score(y_train_overs3,y_pred_train))
print("TEST :",f1_score(y_test,y_pred_test))
```
